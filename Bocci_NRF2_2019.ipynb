```python
import pandas as pd
df = pd.read_csv('bocci_2019_solution.dat', delimiter='\t', header=None, names=[' Model_index', 'Number_of_stable_states',
'Runs_converged_to_steady_state', 'SNAIL', 'miR-34', 'miR-200', 'ZEB', 'E-CAD', 'KEAP1', 'NRF2'])
print(df)
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from sklearn.cluster import KMeans
import numpy as np

# Extract the miR-200 and ZEB columns for clustering
X = df[['miR-200', 'ZEB']].values

# K-means clustering with k=3
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

df['KMeans_Cluster'] = labels

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s =2, alpha = 0.5)
plt.scatter(centers[:, 0], centers[:, 1], marker='x', s=10, linewidths=3, color='red')

# Calculate and plot ellipses for each cluster
for i in range(3):
    # Calculate standard deviation
    std_dev = np.std(X[labels == i], axis=0)

    # Create an ellipse
    ellipse = Ellipse(xy=centers[i], width=2*std_dev[0], height=2*std_dev[1],
                      edgecolor='red', fc='None', lw=2)
    plt.gca().add_patch(ellipse)

plt.xlabel('miR-200')
plt.ylabel('ZEB')
plt.title('K-means Clustering (k=3)')
plt.show()
```

```python
# Map the clusters to the appropriate labels (E, E/M, M)
cluster_map = {
    0: 'E',
    1: 'E/M',
    2: 'M'
}
df['KMeans_Phenotype'] = df['KMeans_Cluster'].map(cluster_map)
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

new_df = df[['E-CAD', 'miR-34', 'miR-200', 'NRF2', 'KEAP1', 'SNAIL', 'ZEB']].copy()

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(-1, 1))
new_df[['E-CAD', 'miR-34', 'miR-200', 'NRF2', 'KEAP1', 'SNAIL', 'ZEB']] = scaler.fit_transform(df[['E-CAD', 'miR-34', 'miR-200', 'NRF2', 'KEAP1', 'SNAIL', 'ZEB']])

print(new_df[['E-CAD', 'miR-34', 'miR-200', 'NRF2', 'KEAP1', 'SNAIL', 'ZEB']])
```

```python
# Perform KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)

# Get cluster labels
clusters = kmeans.fit_predict(new_df)
new_df['Cluster'] = clusters

# Map cluster labels to colors for visualization
cluster_phenotype = {0: 'E', 1: 'E/M', 2: 'M'}
new_df['Phenotype'] = new_df['Cluster'].map(cluster_phenotype)

# Create clustermap
g = sns.clustermap(
    new_df.drop(['Phenotype', 'Cluster'], axis =1),
    method='average',            # Use average linkage method
    metric='euclidean',          # Use Euclidean distance
    cmap='coolwarm',
    figsize=(8, 10),
    col_cluster=False,
    yticklabels=False,
    cbar_pos=(1.05, .2, .03, .4)
)

# Customize the plot
plt.axvline(x=0, color='black', linestyle='--')
plt.text(-0.1, 0.85, 'M cluster', ha='center', va='center', transform=g.ax_row_dendrogram.transAxes)
plt.text(-0.1, 0.5, 'E/M cluster', ha='center', va='center', transform=g.ax_row_dendrogram.transAxes)
plt.text(-0.1, 0.15, 'E cluster', ha='center', va='center', transform=g.ax_row_dendrogram.transAxes)

for label in g.ax_heatmap.get_xticklabels():
    label.set_fontsize(12)
    label.set_fontweight('bold')
plt.show()
```

```python
avg_NRF2 = new_df.groupby('Phenotype')['NRF2'].mean()

print(avg_NRF2)
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram
import seaborn as sns
import matplotlib.pyplot as plt

# Perform KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(df)

# Get cluster centers
cluster_centers = kmeans.cluster_centers_

# Perform hierarchical clustering on the cluster centers
Z = linkage(cluster_centers, method='average', metric='euclidean')

df['Cluster'] = kmeans.labels_

cluster_colors = {0: 'green', 1: 'blue', 2: 'red'}
df['Cluster_Color'] = df['Cluster'].map(cluster_colors)

# Plot the dendrogram
plt.figure(figsize=(8, 5))
dendrogram(Z, labels=['Cluster 1', 'Cluster 2', 'Cluster 3'])
plt.title('Hierarchical Clustering Dendrogram of KMeans Centroids')
plt.xlabel('Cluster')
plt.ylabel('Distance')
plt.show()

# Create clustermap
g = sns.clustermap(
    df[['E-CAD', 'miR-34', 'miR-200', 'NRF2', 'KEAP1', 'SNAIL', 'ZEB']],  # Drop cluster columns for heatmap
    method='average',            # Use average linkage method
    metric='euclidean',          # Use Euclidean distance
    cmap='coolwarm',
    figsize=(8, 10),
    row_colors=df['Cluster_Color'],
    col_cluster=False,
    yticklabels=False,
    cbar_pos=(1.05, .2, .03, .4)
)

plt.show()
```
